```python
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

```python
ferry_df = pd.read_csv('nys_ferry/Monthly_Interstate_Ferry_Ridership__Port_Authority___Beginning_1998.csv')
```

```python
ferry_df.head()
```

```python
ferry_df.describe()
```

```python
from datetime import datetime

for i in ferry_df['Month']:
    d = datetime.strptime(i.strip(), '%b')
    i = d.strftime('%m')

ferry_df['Month'] = ferry_df['Month'].map(lambda x: (datetime.strptime(x.strip(), '%b') ).strftime('%m'))
ferry_df['month_year'] = pd.to_datetime(ferry_df['Month'].astype(str) + ' ' + ferry_df['Year'].astype(str))
```

```python
ferry_df.sort_values(by='month_year',inplace = True)
```

```python
plt.figure(figsize=(15,10))
plt.plot(ferry_df['month_year'],ferry_df['Downtown Passengers'])
```
