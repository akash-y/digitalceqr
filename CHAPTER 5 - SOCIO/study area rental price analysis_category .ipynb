### Data downloaded from streeteasy, and first plot median asking rent for studio 

## Studio

```python
studio = pd.read_csv('medianAskingRent_Studio.csv')
studio.head()
```

#### Data contains rental price in many area, drop other area and keep the study area. There are many nan in RedHook data, so we also drop it. 

```python
studio_study= studio.loc[oneb['areaName'] == 'Red Hook']
studio_study2 = studio_study.dropna(axis=1, how="all")
```

```python
studio_study3=studio_study2.drop(['Borough','areaType'], axis=1)
studio_study4 = studio_study3.set_index(['areaName'])
studio_study4
```

```python
df_studio= studio_study4.T
```

```python
df = df_studio.rename(columns={'areaName': 'year', 'Red Hook': 'medianAskingRent'})
```

```python

ax =df [['medianAskingRent']].plot(kind='bar', title ="median asking rental price in Redhook (Studio)", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("median asking rental price", fontsize=12)
plt.show()
```

## One-Bedroom

```python
oneb = pd.read_csv('medianAskingRent_Onebd.csv')
oneb.head()
```

```python
oneb_study= oneb.loc[oneb['areaName'] == 'Red Hook']
oneb_study2 = oneb_study.dropna(axis=1, how="all")
```

```python
oneb_study3=oneb_study2.drop(['Borough','areaType'], axis=1)
oneb_study4 = oneb_study3.set_index(['areaName'])
oneb_study4
```

```python
df_oneb= oneb_study4.T
```

```python
df1
```

```python
ax =df1[['medianAskingRent']].plot(kind='bar', title ="median asking rental price in Redhook (Onebedroom)", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("median asking rental price", fontsize=12)
plt.show()
```

```python

f, ax = plt.subplots(figsize=(20,8))
df1['medianAskingRent'].plot(alpha=1,linewidth=2,ax=ax,label='medianAskingRent')
plt.title("median asking rental price in Redhook (Onebedroom)", fontsize=25)
plt.xlabel("Year", fontsize=15)
plt.ylabel("median asking rental price", fontsize=15)
plt.legend(prop={'size': 20})
plt.show()
```
